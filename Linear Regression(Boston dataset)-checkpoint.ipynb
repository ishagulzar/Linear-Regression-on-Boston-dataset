## Using the famous Boston Housing Dataset for Single Linear Regression

### Importing relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# Setting random seed for reproducibility
# By setting the seed value to a specific number, such as 0 in this case, you ensure that the sequence of random numbers generated by NumPy will be the same every time you run the code.
# This can be helpful when you want to debug or test your code and need consistent results.

np.random.seed(0)
```

```python
# Reading the dataset
df = pd.read_csv("boston.csv")
df
```

```python
# Name of the columns of our dataset 
df.columns
```

```python
# Splitting the dataset into features(X) and target variable(y)

X = df.iloc[:,:-1]
y = df.MEDV
```

```python
# Splitting dataset into training and test sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# Dimensions of training and test sets

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Calculating the correlation coefficient between different features in the training and the target variable

X_train.CRIM.corr(y_train), X_train.ZN.corr(y_train), X_train.INDUS.corr(y_train), X_train.CHAS.corr(y_train), X_train.NOX.corr(y_train), X_train.RM.corr(y_train), X_train.AGE.corr(y_train), X_train.DIS.corr(y_train), X_train.RAD.corr(y_train), X_train.TAX.corr(y_train), X_train.PTRATIO.corr(y_train), X_train.B.corr(y_train), X_train.LSTAT.corr(y_train)
```

```python
plt.scatter(X_train.RM, y_train)
```

```python
# fitting the model
lm = LinearRegression()
model_slr = lm.fit(X_train[['RM']], y_train)
```

```python
#---------------------------------------------#
```

```python
model_slr.intercept_ , model_slr.coef_
```

```python
model_slr.predict(X_train[['RM']])
```

```python
# compute predictions using the model parameters
yhat_train = model_slr.intercept_ + model_slr.coef_ * X_train['RM']
yhat_train[0:5]
```

```python
# calculate R Square = 1 - SSE/SST

# sum of squares of errors
SSE = sum((y_train - yhat_train)**2)
```

```python
# total sum of squares or total variation in target
SST = sum((y_train - np.mean(y_train))**2)
```

```python
# R Square
model_slr_r_square = 1 - SSE/SST
model_slr_r_square
```

```python
# sum of squares of regression (variation explained by the model)
SSR = sum((yhat_train - np.mean(y_train))**2)
```

```python
SSR/SST
```

```python
# total variation = explained variation + unexplained variation
# SST = SSR + SSE
SST, SSR + SSE
```

```python
# get this using the model
model_slr.score(X_train[['RM']],y_train)
```

```python
# correlation between feature and label (r)
r = X_train.RM.corr(y_train)
r
```

```python
# this is almost as the model R Square. Shows that as correlation is higher model fit is better.
r**2
```

```python
# plot the model
plt.scatter(X_train.RM,y_train)
plt.plot(X_train.RM,yhat_train,c='r')
plt.show()
```

```python
# apply model to predict test data
yhat_test = model_slr.predict(X_test[['RM']])
```

```python
plt.scatter(X_test.RM,y_test)
plt.plot(X_test.RM,yhat_test,c='r')
plt.show()
```

```python
# compute train and test MSE
mse_train = np.mean((y_train - yhat_train)**2)
mse_test = np.mean((y_test - yhat_test)**2)
mse_train,mse_test
```

```python
# compute train and test MAPE
mape_train = np.mean(abs(y_train - yhat_train)/y_train) * 100
mape_test = np.mean(abs(y_test - yhat_test)/y_test) * 100
mape_train,mape_test
```

```python
model_slr_performance = {'model':"model_slr",'R Sq':model_slr_r_square,'train_MSE':mse_train,'test_MSE':mse_test,'train_MAPE':mape_train,'test_MAPE':mape_test}
```

```python
df_snapshot = pd.DataFrame(columns=['model','R Sq','train_MSE','test_MSE','train_MAPE','test_MAPE'])
df_snapshot
```

```python
df_snapshot = df_snapshot.append(model_slr_performance,ignore_index=True)
```
